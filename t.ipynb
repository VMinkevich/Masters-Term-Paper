```python
import numpy as np
import pandas as pd
from scipy.stats import norm
import random
import time
import matplotlib.pyplot as plt
```

```python
# Заданные параметры
lambda_poisson = 5  # Параметр распределения Пуассона
mu_normal = 0       # Математическое ожидание для нормального распределения
sigma_normal = 1    # Стандартное отклонение для нормального распределения
num_samples = 100000 # Количество выборок для Монте-Карло

# Генерация случайных величин
X = np.random.poisson(lambda_poisson, num_samples)   # Случайная величина с распределением Пуассона
Y = np.random.normal(mu_normal, sigma_normal, num_samples)  # Случайная величина с нормальным распределением

# Сумма случайных величин
Z = X + Y

# Построение гистограммы для суммы Z
plt.figure(figsize=(10, 6))
plt.hist(Z, bins=50, density=True, alpha=0.6, color='g', label='Сумма Z')
plt.hist(X, bins=50, density=True, alpha=0.6, color='b', label='Пуассон')
plt.hist(Y, bins=50, density=True, alpha=0.6, color='r', label='Нормальное')

# Настройка графика
plt.title('Распределение сумма двух независимых случайных величин (Пуассон + Нормальное)', fontsize=14)
plt.legend(loc='upper right', fontsize=12)
plt.xlabel('Значение суммы Z', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)

# Показать график
plt.show()
```
